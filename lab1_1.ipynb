Імпорт модулів та задання шляху до файлу

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "spotify-2023.csv"
data = pd.read_csv(file_path, encoding='latin1')
```

Побудова горизонтальної стовпчикової діаграми для 10 найпопулярніших пісень на Spotify у 2023 році за кількістю прослуховувань:


```python
top = data.head(10)
plt.figure(figsize=(10, 6))
plt.barh(top['track_name'], top['streams'], color='blue')
plt.xlabel('Кількість прослуховувань')
plt.ylabel('Назва треку')
plt.title('10 найпопулярніших пісень на Spotify у 2023 році')
plt.gca().invert_yaxis()
plt.show()
```

Розподіл найпопулярніших виконавців пісень: Створіть кругову діаграму або
секторну діаграму, яка відображає розподіл виконавців серед
найпопулярніших пісень у 2023 році.

```python
artist_distribution = data['artist(s)_name'].value_counts().head(10)

# Побудова кругової діаграми
plt.figure(figsize=(8, 8))
plt.pie(artist_distribution, labels=artist_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Розподіл найпопулярніших артистів 2023')
plt.show()
```

. Рейтинг артистів на Spotify: Створіть стовпчикову діаграму, на якій
відображаються 10 найпопулярніших артистів на Spotify у 2023 році за
кількістю слухачів.

```python
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')
top_artists = data.groupby('artist(s)_name')['streams'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
top_artists.plot(kind='bar', color='red')
plt.xlabel('Виконавець')
plt.ylabel('Кількість прослуховувань')
plt.show()

```

Порівняння популярності трьох треків: Побудуйте горизонтальну
стовпчикову діаграму, яка порівнює кількість прослуховувань трьох
конкретних треків у 2023 році.

```python
tracks = data.iloc[[0, 1, 2]]

plt.figure(figsize=(10, 6))
plt.barh(tracks['track_name'], tracks['streams'], color=['green', 'blue', 'red'])
plt.xlabel('Прослуховувань (МЛН)')
plt.ylabel('Назва треку')
plt.title('Порівняння трьох треків')
plt.show()
```
