```python
import numpy as np
import re
import tensorflow as tf
from tensorflow.keras.layers import Dense, LSTM, Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

texts_positive = [
    "Священна і правдива річ, Воля України!",
    "Ти диво, ти країно моя! Ти земле, ти світе мій!",
    "За країну нашу святую гинуть наші браття."
]

texts_negative = [
    "Сумно зриваються пісні дні",
    "Вітер віє, вітрила рве, відпливає корабель.",
    "Не коси, не коси, чорнобривії, та ти спиш і засинаєш."
]

texts = texts_positive + texts_negative
count_positive = len(texts_positive)
count_negative = len(texts_negative)

max_words_count = 1000
tokenizer = Tokenizer(num_words=max_words_count, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n\r«»', lower=True, split=' ', char_level=False)
tokenizer.fit_on_texts(texts)

max_text_len = 10
data = tokenizer.texts_to_sequences(texts)
data_pad = pad_sequences(data, maxlen=max_text_len)

X = data_pad
Y = np.array([[1, 0]] * count_positive + [[0, 1]] * count_negative)

indices = np.random.choice(X.shape[0], size=X.shape[0], replace=False)
X = X[indices]
Y = Y[indices]

model_lstm = Sequential()
model_lstm.add(Embedding(max_words_count, 128, input_length=max_text_len))
model_lstm.add(LSTM(128, return_sequences=True))
model_lstm.add(LSTM(64))
model_lstm.add(Dense(2, activation='softmax'))
model_lstm.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer=Adam(0.0001))

history_lstm = model_lstm.fit(X, Y, batch_size=32, epochs=50)
```
